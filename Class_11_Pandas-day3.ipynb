```python
import pandas as pd
```

```python
df_sales = pd.read_csv('D:/Pandas_sample/sales_data_final.csv')
```

```python
df_sales.head()
```

```python
def profit_flag(a):
    if  a>= -0:
        return 'Positive'
    else:
        return 'Negative'
```

```python
profit_flag(-25)
```

```python
profit_flag(15)
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a):
    if a < 10:
        return 'Low'
    elif a > 10 and a < 20:
        return 'Medium'
    else:
        return 'High'
```

```python
df_sales['quantity_flag'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a**2)
```
